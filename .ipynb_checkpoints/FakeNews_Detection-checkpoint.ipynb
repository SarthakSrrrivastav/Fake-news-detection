```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
import re  # use for searching word in a text or paragraph
from nltk.corpus import stopwords                               # remove the stopwords
from nltk.stem.porter import PorterStemmer                      # remove the prefix and sufix
from sklearn.feature_extraction.text import TfidfVectorizer     # covverts words into Vector
```

```python
import nltk
nltk.download('stopwords')
```

```python
df=pd.read_csv('E:/JUPYTER/ML Projects/Fake News Detection/fakenews.csv')
```

```python

df=df.fillna('')   # replacing null value with empty string
```

```python
df.isnull().sum()
```

```python
df['content']=df['author']+' '+ df['title']  # creating new column called CONTENT and combinig the AUTHOR nad Title column
```

```python
df['content']
```

```python
Y=df['label']
X=df.drop(columns='label',axis=1)
```

```python
df
```

## Stemming

```python
port_stem=PorterStemmer()
```

```python
def stemming(content):  # in stemming paarameter its not compulsory to give that name , we can give any name its a input for  text
    
    
    stemmed_content=re.sub('[^a-zA-Z]',' ',content) # it will remove all the puntuation,number,special char and keep only albhabets CAPTIAL and small  and  NUMBER and PUNTUATION will be replaced by empty string
    stemmed_content=stemmed_content.lower()
    stemmed_content=stemmed_content.split()
    stemmed_content=[port_stem.stem(word) for word in stemmed_content if not word in stopwords.words('english')]
    stemmed_content=' '.join(stemmed_content)
    return stemmed_content
```

```python
df['content']=df['content'].apply(stemming)
```

```python
df['content']
```

```python
X=df['content']
```

```python
Y=df['label']
```

```python
X
```

## Coverting Textual data into Vector form

```python
vectorizer=TfidfVectorizer()
vectorizer.fit(X)
X=vectorizer.transform(X)
```

```python
print(X)
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=1)
```

## Training Model

```python
model=LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

## Checking Accuracy Score

```python
# Checking accuracy score on training data

X_train_preddiction=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_preddiction,Y_train)
training_data_accuracy
```

```python
# Checking accuracy on test data
X_test_prediction=model.predict(X_test)
test_data_accuracy=accuracy_score(X_test_prediction,Y_test)
test_data_accuracy
```

## Predective System

```python
X_news=X_test[0]

prediction=model.predict(X_news)
print(prediction)
if (prediction==0):
    print('News is Real')
else:
    print('News is Fake')
```
